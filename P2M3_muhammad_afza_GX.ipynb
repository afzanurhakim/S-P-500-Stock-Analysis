# Data Validation with Great Expectation

# Import Libraries

```python
import pandas as pd
import great_expectations as gx
```

# Great Expectation

```python
df = pd.read_csv("P2M3_muhammad_afza_data_clean.csv")
df
```

```python
# buat context
context = gx.get_context()

datasource = context.data_sources.add_pandas(name="sp500_datasource")

# tambahkan csv
asset = datasource.add_csv_asset(
    name="sp500_asset",
    filepath_or_buffer="P2M3_muhammad_afza_data_clean.csv"
)

# buat batchrequest
batch_request = asset.build_batch_request()

# buat validator
context.suites.add(gx.ExpectationSuite(name="sp500_suite"))
validator = context.get_validator(batch_request=batch_request, expectation_suite_name="sp500_suite")
```

# Expectation 1 - To be Unique

```python
# memastikan kolom 'symbol' unik untuk identifikasi ticker
validator.expect_column_values_to_be_unique(column='symbol')
```

# Expectation 2 - To be Between

```python
# memastikan harga saham (price) berada di atas 0
validator.expect_column_values_to_be_between(column='price', min_value=0.01, max_value=5000)
```

# Expectation 3 - To be in Set

```python
# mengambil unique value acuan 'set'
valid_symbols = df['symbol'].unique().tolist()

# memvalidasi
validator.expect_column_values_to_be_in_set('symbol', valid_symbols)
```

# Expectation 4 - To be in Type List

```python
# memastikan kolom market_cap memiliki tipe data numerik (float atau integer) untuk kalkulasi finansial
validator.expect_column_values_to_be_in_type_list('market_cap', ['float64', 'int64'])
```

# Expectation 5 - Column Value Lengths to be Between

```python
# memvalidasi bahwa panjang karakter di kolom 'symbol' berada dalam rentang 1 sampai 5
validator.expect_column_value_lengths_to_be_between('symbol', min_value=1, max_value=5)
```

# Expectation 6 - Column Values to Match Regex

```python
# memastikan kolom 'symbol' berisi huruf kapital
validator.expect_column_values_to_match_regex('symbol', r'^[A-Z.\-]+$')
```

# Expectation 7 - Column Count to Equal

```python
# memastikan jumlah total kolom di dalam tabel sesuai dengan dataset
validator.expect_table_column_count_to_equal(15)
```
